```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import louvain_communities
```

```python
#Communautés nettes
# Ils parlent beaucoup entre eux et Ils parlent très peu aux autres
# commu floue : Ils parlent entre eux MAIS ils parlent presque autant aux autres groupes

G = nx.planted_partition_graph(4, 5, 0.8, 0.1, seed=42)
# La valeur 42 est arbitraire juste pour pourvoir reproduire les mm résultats
# assurer la reproductibilité

communautes = louvain_communities(G, seed=42)
mod = nx.community.modularity(G, communautes)

print("Modularite:", round(mod, 3))
print("Communautes trouvees:", len(communautes))

partition = {}
for i, comm in enumerate(communautes):
    for noeud in comm:
        partition[noeud] = i

pr = nx.pagerank(G)
tailles = [3000 * pr[n] for n in G.nodes()]
couleurs = [partition[n] for n in G.nodes()]

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, node_color=couleurs, node_size=tailles,
        cmap=plt.cm.tab10, edge_color='gray', with_labels=True)

plt.title(f"Communautes nettes - Modularite: {mod:.3f}")
plt.show()
```
