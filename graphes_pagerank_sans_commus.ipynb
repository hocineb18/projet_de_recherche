```python
G = nx.erdos_renyi_graph(20, 0.1, seed=42)

communautes = louvain_communities(G, seed=42)
mod = nx.community.modularity(G, communautes)

print("Modularite:", round(mod, 3))
print("Communautes trouvees:", len(communautes))

pr = nx.pagerank(G)
valeurs_pr = [pr[n] for n in G.nodes()]

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)

nx.draw_networkx_edges(G, pos, edge_color='gray')

nodes = nx.draw_networkx_nodes(
    G, pos,
    node_color=valeurs_pr,
    node_size=500,
    cmap=plt.cm.viridis
)

nx.draw_networkx_labels(G, pos)

plt.colorbar(nodes, label="PageRank")
plt.title(f"Graphe aleatoire - Modularite: {mod:.3f}")
plt.show()
```

```python
G = nx.planted_partition_graph(4, 5, 0.4, 0.3, seed=42)
# Une communauté floue = un groupe de nœuds où la séparation entre les groupes n’est pas nette.
# ça se joue au niveau des probabilités dans la fct planted_partition_graph 

communautes = louvain_communities(G, seed=42)
mod = nx.community.modularity(G, communautes)

print("Modularite:", round(mod, 3))
print("Communautes trouvees:", len(communautes))

pr = nx.pagerank(G)
valeurs_pr = [pr[n] for n in G.nodes()]

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)

nx.draw_networkx_edges(G, pos, edge_color='gray')

nodes = nx.draw_networkx_nodes(
    G, pos,
    node_color=valeurs_pr,
    node_size=250,
    cmap=plt.cm.viridis
)

nx.draw_networkx_labels(G, pos)

plt.colorbar(nodes, label="PageRank")
plt.title(f"Communautes floues - Modularite: {mod:.3f}")
plt.show()
```
