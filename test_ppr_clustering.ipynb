# Test du clustering PPR
Tests de l'algorithme de détection de communautés basé sur Personalized PageRank.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from community_detection_robust_avec_virtual_community import ppr_clustering
```

## 1. Création d'un graphe test
Deux communautés reliées par une seule arête.

```python
G = nx.Graph()

# communauté 1
G.add_edges_from([(0,1),(1,2),(0,2)])

# communauté 2
G.add_edges_from([(3,4),(4,5),(3,5)])

# lien entre communautés
G.add_edge(2,3)

A = nx.to_numpy_array(G)

print(A)
```

## 2. Visualisation du graphe

```python
nx.draw(G, with_labels=True)
plt.show()
```

## 3. Lancer l'algorithme de clustering

```python
history = ppr_clustering(A)

for level, communities in enumerate(history):
    print(f"Niveau {level}:", communities)
```

resultat attendu : 

Niveau 0: [{0}, {1}, {2}, {3}, {4}, {5}]

Niveau 1: [{0,1,2}, {3}, {4}, {5}]

Niveau 2: [{0,1,2}, {3,4,5}]

Niveau 3: [{0,1,2,3,4,5}]
